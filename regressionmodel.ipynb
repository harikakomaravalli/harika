```python
#split the data into training/testing
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#divide your data into dependent and independnt variable(s)
x = boston_data.iloc[:,:-1]
y = boston_data.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train.shape
x_test.shape
```

```python
#building the model
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
#create a table showing the coefficient (influence)
#of each of the columns
cdf = pd.DataFrame(lin_reg.coef_,index=d.feature_names,
                   columns=['Coefficients'])
cdf
```

```python
#featureScaling is to be done to make every column standard
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_tarin,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x_train,y_train)
```

```python
cdf = pd.DataFrame(lin_reg.coef_,index=d.feature_names,columns=['Coefficients'])
cdf
```

```python
#then we go for prediction and calculate the metrics
reg_pred = lin_reg.predict(x_test)
```

```python
#r-squared value
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,reg_pred))
print(r2_score(y_test,reg_pred)*100)
```

```python
#once we are confident enough to test the model in production-ready environment
#will save the models using pickle
import pickle
```

```python
#we will create pickle files for both regression model and scaling model
#open() is generally used for file modes
pickle.dump(lin_reg,open('regmodel.pkl','wb'))#wb=write binary  format and we will use 'w' for text file 
```
